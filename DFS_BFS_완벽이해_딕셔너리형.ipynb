# DFS BFS 기본 원리 및 코드

DFS는 깊이 우선 탐색, BFS는 너비 우선 탐색3

1) DFS는 최대한 깊이 내려간 뒤, 더이상 깊이 갈 곳이 없을 경우 옆으로 이동

일반적으로 DFS는 스택 또는 재귀함수로 구현


그래프 : node들과 이 node들을 연결하는 간선으로
이루어진 자료구조를 말함.


그래프 탐색은 하나의 정점을 시작으로 다른 정점들을 모두 한번씩 방문하는 것을 말함.


1. 입력값 저장
우선 DFS, BFS를 구현하기 위해선 입력값을 queue에 저장하는 단계가 필요.

```python
# BFS 너비 우선 탐색

def bfs(graph, start_node): # 'A'
    visit = list()
    queue = list() # 노드를 선입선출
    
    queue.append(start_node) # queue = ['A']
    print(queue)
    
    while queue: # 큐의 목록이 바닥나, 더이상 방문할 노드가 없을 때까지~
        node = queue.pop(0) # node = A
        print("node는 =",node)
        if node not in visit: # 방문 X
            visit.append(node) # visit= ['A'] 방문 기록 남기고,
            print("방문한 visit : ",visit,"\n")
            queue.extend(graph[node])# 그래프 값에 접근한 걸 추가
            print("이 노드와 연결되어 있는",queue)
        print("한개의 노드 탐색 끝",node,"\n")
# append는 리스트 끝에 추가
# extend는 리스트 끝에 가장 바깥쪽 iterable(반복가능한,객체)의 모든 항목을 넣는다.
    
    return visit
```

```python
# DFS 깊이 우선 탐색 - 기억하자. BFS는 큐를, DFS는 스택을 사용
# 스택으로 후입후출이다보니, 더 깊게만 들어가고, 앞에 있는
# 너비 애들의 탐색 순서는 계속 뒤로 밀림.
def dfs(graph,start_node):
    visit = list()
    stack = list() # 노드를 선입후출
    
    stack.append(start_node)
    print(stack)
    
    while stack:
        node = stack.pop()
        print("node는 =",node)
        if node not in visit:
            visit.append(node)
            print("방문한 visit : ",visit,"\n")
            stack.extend(graph[node])
            print("이 노드와 연결되어 있는",stack)
        print("한개의 노드 탐색 끝",node,"\n")
            
```

```python
dfs(graph,'A')
```

```python
import sys
from collections import deque # 큐

sys.setrecursionlimit(10 ** 8) # 백준에서 탐색 메모리 더 확보하려고

n, m, start = map(int, input().split(' '))
# 첫째에 노드의 개수, 둘째에 간선의 갯수, 셋째로 탐색 시작할 노드 번호 
# n = 4 , m = 5 , start = 1 
g = [[] for i in range(n + 1)]

for i in range(1, m + 1):
    a, b = map(int, input().split(' '))
    g[a].append(b)
    g[b].append(a)

visited_dfs = [False] * (n + 1)
visited_bfs = [False] * (n + 1)

print(g)

for i in range(1, n + 1):
    g[i].sort()

print(g)


def dfs(s):
    print(s, end=" ")
    visited_dfs[s] = True

    for i in g[s]:
        if not visited_dfs[i]:
            dfs(i)

def dfs_stk(s):
    stk = [s]

    while stk:
        nn = stk.pop()
        if not visited_dfs[nn]:
            print(nn, end=' ')
            visited_dfs[nn] = True
            # stk 에 넣기 때문에 역순으로 append
            for c in g[nn][::-1]:
                if not visited_dfs[c]:
                    stk.append(c)

def bfs(s):
    visited_bfs[s] = True
    q = deque([s])

    while q:
        p = q.popleft()
        print(p, end=" ")
        for i in g[p]:
            if not visited_bfs[i]:
                q.append(i)
                visited_bfs[i] = True

dfs_stk(start)
print('')
bfs(start)
```
